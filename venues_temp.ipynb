```python
df_venues['class1'] = df_venues.apply (lambda row: row['Category Class'].split('/')[-2], axis=1)
df_venues.head(10)
```

```python
df_venues['class2'] = df_venues.apply (lambda row: (row['Category Class'].split('/')[-2]+' '+row['Category Class'].split('/')[-1]), axis=1)
df_venues.head(10)
```
